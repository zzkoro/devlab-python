# 시카고 샌드위치 맛집

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.parse import urljoin
import re

url_base = "http://www.chicagomag.com"
url_sub = "/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/"
url = url_base + url_sub

html = urlopen(url)

soup = BeautifulSoup(html, "html.parser")
```

```python
sammy = soup.select('div.sammy')

rank = []
main_menu = []
cafe_name = []
url_add = []

for item in sammy:
    sammyRank = item.find(class_='sammyRank').get_text()
    sammyListing = item.find(class_='sammyListing').get_text()
    sammyHref = item.find('a')['href']
    
    rank.append(sammyRank)
    main_menu.append(re.split(('\n|\r\n'), sammyListing)[0])
    cafe_name.append(re.split(('\n|\r\n'), sammyListing)[1])

    url_add.append(urljoin(url_base, sammyHref))
```

```python
len(rank), len(main_menu), len(cafe_name), len(url_add)
```

```python
import pandas as pd

data = {'Rank': rank, 'Menu':main_menu, 'Cafe':cafe_name, 'URL':url_add}
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('./data/best_sandwiches_list_chicago.csv', sep=',', encoding='UTF-8')
```

```python
from tqdm import tqdm_notebook
price = []
address = []

for n in tqdm_notebook(df.index):
    html = urlopen(df['URL'][n])
    soup_tmp = BeautifulSoup(html, 'lxml')
    gettings = soup_tmp.find('p', 'addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

```python
price
```

```python
address
```

```python
len(price), len(address), len(df)
```

```python
df['Price'] = price
df['Address'] = address
df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address']]
df.set_index('Rank', inplace=True)
```

```python
df.head()
```

```python
df.to_csv('./data/best_sandwiches_list_chicago2.csv', sep=',', encoding='UTF-8')
```
