```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('./data/salesfunnel.xlsx')
df.head()
```

```python
pd.pivot_table(df, index=["Name", "Rep"], values=["Price"], aggfunc=np.sum)
```

```python
pd.pivot_table(df, index=["Manager", "Rep", "Product"],
              values=["Price", "Quantity"],
              aggfunc=[np.sum, np.mean],
              fill_value=0,
              margins=True)
```
